- Feladata: hosszú formátúvá alakítja az adattáblát, úgy, hogy a kutyák és futás/váltás adataik egymás alatt helyezkednek el.
- Külön figyelve arra, hogy mindegyik kutyának megmaradjanak a sorban a váltójára és a versenyre jellemző közös adatoszlopok.
- Ebben az adatbázisban van még két kód az elején, feladatuk:
- a W/L/T oszlop hiányzó értékek számolása
- kék/piros pálya oszlop szín kiosztása

```python
import pandas as pd
#adatbázis betöltése
flyball2 = pd.read_csv('./../data/flyballdf.csv', index_col=[0])
```

```python
flyball2.isna().sum()
```

```python
#A W/L/T oszlop  107 hiányzó értékének számítása: NaN értékeknek a 'NON' stringet adjuk
flyball2['W/L/T'].fillna('NON', inplace=True)
```

```python
#A NON értékek feltöltése egy elemző kóddal történik: eldönti a Total time idő alapján, hogy ki nyert 
for i in range(len(flyball2)):
    if flyball2.iloc[i,5] == 'NON':                         #vizsgálja, hogy a NON sorokban
        if flyball2.iloc[i,1] == flyball2.iloc[i+1,1]:     #ha megegyezik a WHEN dátuma
            if flyball2.iloc[i,4] > flyball2.iloc[i+1,4]:    #akkor a TOTAL TIME hol nagyobb, hol kisebb
                flyball2.iloc[i,5] = 'L'
            if flyball2.iloc[i,4] < flyball2.iloc[i+1,4]:
                flyball2.iloc[i,5] = 'W'
                
            if flyball2.iloc[i,4] > flyball2.iloc[i+1,4]:    #és eszerint átírja a NON-t W/L-re
                flyball2.iloc[i+1,5] = 'W'
            if flyball2.iloc[i,4] < flyball2.iloc[i+1,4]:
                flyball2.iloc[i+1,5] = 'L'
            
            if flyball2.iloc[i,4] == flyball2.iloc[i+1,4]:
                flyball2.iloc[i,5] = 'T'
                flyball2.iloc[i+1,5] = 'T'
```

```python
len(flyball2[flyball2['W/L/T'] == 'NON'])
#a feltöltés után 25 sor marad, aminek nincs NON időpárja, de valószínűleg többségük kiesik a tisztítás során
```

```python
#a place, date és competition oszlopokat az elejére kell mozgatni, hogy közös adatknék a váltó mindegyik kutyájánál lássuk
firstplace = flyball2.pop('place')
flyball2.insert(2 ,'place', firstplace)

firstdate = flyball2.pop('date')
flyball2.insert(2 ,'date', firstdate)

firstcompetition = flyball2.pop('competition')
flyball2.insert(2 ,'competition', firstcompetition)

flyball2.head(150)
```

```python
flyball2.reset_index(inplace = True, drop = True) #újraindexelve
flyball2.shape
```

```python
#ide jöhetnek a kék-piros kategória oszlop értékeinek számítása:

flyball2['colour'] = 0
for i in range(len(flyball2)-1):
    
    if flyball2.iloc[i, 1] == flyball2.iloc[i+1, 1]:
        
        flyball2.iloc[i, 22] = 'Blue'
        
        flyball2.iloc[i+1, 22] = 'Red'


len(flyball2[flyball2['colour'] == 0])/len(flyball2) #1,2% nincs kategorizálva, ellenőrizzük tisztitás után
```

```python
#mozgassuk a colour oszlopot is az elejére, mert ott vannak azok az információk, amiket minden kutyához 
#rendelni szeretnék

firstcolour = flyball2.pop('colour')
flyball2.insert(2 ,'colour', firstcolour)
```

```python
#4 új oszlop beszúrása
flyball2.insert(loc=11, column='rank', value='1')
flyball2.insert(loc=15, column='rank2', value='2')
flyball2.insert(loc=19, column='rank3', value='3')
flyball2.insert(loc=23, column='rank4', value='4')
```

```python
#a színes adatbázisok kijelölése
feherdf = flyball2.iloc[0:,0:11]
pirosdf = flyball2.iloc[0:,11:15]
kekdf = flyball2.iloc[0:,15:19]
sargadf = flyball2.iloc[0:,19:23]
zolddf = flyball2.iloc[0:,23:27]
```

```python
#a színes adatbázisok újrafűzése, oszlopnevek egységesítése
pirosdf.columns = ['rank', 'name', 'relay', 'time']
kekdf.columns = ['rank', 'name', 'relay', 'time']
sargadf.columns = ['rank', 'name', 'relay', 'time']
zolddf.columns = ['rank', 'name', 'relay', 'time']

feherpirosdf = pd.concat([feherdf, pirosdf], axis=1)
feherkekdf = pd.concat([feherdf, kekdf], axis=1)
fehersargadf = pd.concat([feherdf, sargadf], axis=1)
feherzolddf = pd.concat([feherdf, zolddf], axis=1)

szines2 = pd.concat([feherpirosdf, feherkekdf, fehersargadf, feherzolddf], axis=0)
szines2.reset_index(inplace = True, drop = True) #eldobja a régit
```
